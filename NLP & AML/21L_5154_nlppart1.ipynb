```python
import re

def get_words(text):
 pattern1=r'\b[Zz][a-z]*a\b'
 w1=re.findall(pattern1, text)
 if not w1:
  word1=None
 else:
  word1 = w1[0]

 pattern2 = r'\b[0-9]k[a-z]*[0-9]\b'
 sec_word = re.search(pattern2, text)
 if sec_word is None:
  w2 = None
 else:
  w2 = sec_word.group()
 if w2 is None:
  word2=None
 else:
  word2 = w2[3:-3]

 pattern3 = r'\bc[a-z]*[*][a-z]+a\b'
 third_word = re.search(pattern3, text)
 if third_word is None:
  w3 = None
 else:
  w3 = third_word.group()
 if w3 is None:
  word3 = None
 else:
  word3 = w3.replace('*', '')

 pattern4 = r'\b[pP][a-z]{2}a\b'
 w4 = re.findall(pattern4, text)
 if w4 is None:
  word4 = None
 else:
  word4 = w4[0]
 return word1, word2, word3, word4, len(w4), len(w1)


text="""Pila Forfeited you engrossed but 1kometimes explained. Another 1kacokaco1 as studied it to evident.
Merry sense 9given he be arisepila. Conduct at an replied removal an amongst.
Remainingzalima 0determine few her two cordially Zalima admitting old. Sometimes ctra*nger his
pisdsdla ourselves her co*la depending you boy. Eat discretion cultivated possession far comparison
projection pila considered. And few fat interested discovered inquietude insensible unsatiable increasing
zalima eat."""
word1,word2,word3,word4,l1,l2=get_words(text)
output = f"{word1} {word2} {word3} {word4} de"

print("First word:",word1, " and Count:",l1)
print("Second word:",word2)
print("Third word:",word3)
print("Fourth word:",word4, " and Count:",l2)
print("Final Message:",output)
```

```python
import nltk, re, pprint
from urllib import request
from bs4 import BeautifulSoup
from nltk.stem import PorterStemmer, LancasterStemmer
from nltk.tokenize import word_tokenize

#q1
def pattern_reg(t):
    patt= {
        'a': r'[a-zA-Z]+',
        'b': r'[A-Z][a-z]*',
        'c': r'p[aeiou]{,2}t',
        'd': r'\d+(\.\d+)?',
        'e': r'([^aeiou][aeiou][^aeiou])*',
        'f': r'\w+|[^\w\s]+'
    }
    for key, pattern in patt.items():
        print(f"Pattern {key}: {pattern}")
        nltk.re_show(pattern, t)
        print("\n")
#q2
def determiner(t):#getting text and applying
    p=r'\b(?:a|an|the)\b'#oly matching lower case one as mentioned in ass(a,an,the)
    result=re.findall(p, t)
    return result

def ari_exp(t):
    #p=r'\b\d+[+*]\d+(?:[+*]\d+)*\b'#captures + or *
    p=r'\b\d+\*\d+[+]\d+(?:[+*]\d+)*\b|\b\d+[+]\d+\*\d+(?:[+*]\d+)*\b' #captures + and * input only
    result=re.findall(p, t)
    return result
#q5
#porter & lancaster stemmer to normalize some tokenized txt
def normalize_text(t):
    p=PorterStemmer()
    l=LancasterStemmer()
    words=word_tokenize(t)
    plist=[]
    Llist=[]
    for w in words:
      plist.append(p.stem(w))#result of applying porter stemming
    for w in words:
      Llist.append(l.stem(w)) #result of applying lan stemming
    return plist, Llist

in1="Pila forfeited you are engrossed but 1kometimes 1245 explained. pit, pot, parrot."
pattern_reg(in1)

in2="A quick brown fox jumps over the lazy The An dog. An apple a day keeps the doctor away."
print("Determiners:",determiner(in2))
#i made sure only input conatining both * and + are captured only like first 3 and last one
in3="2*3+8 4+5*6 10*10+1 8 8+ 8*0 8+0 8+0* 8+0*4"
print("Arithmetic Expressions:",ari_exp(in3))
print('\n')
'''
Porter tends to preserve more of the original word while
Lancaster is more aggressive and produces shorter stems, sometimes at the cost of readability.
'''
porter_stems, lancaster_stems=normalize_text(in1)
print("Porter Stems:",porter_stems[:10])
print("Lancaster Stems:",lancaster_stems[:10])
```
