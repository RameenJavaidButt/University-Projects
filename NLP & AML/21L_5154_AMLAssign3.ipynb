```python
import matplotlib.pyplot as plt

X1=[]
y1=[]

with open('ex1data1.txt', 'r') as file:
    for line in file:
        a=line.strip().split(',')#getting the values form file
        X1.append(float(a[0]))
        y1.append(float(a[1]))

m=len(y1)#noof training examples
plt.figure(figsize=(8,5))
plt.plot(X1, y1, 'rx', markersize=5)
plt.xlabel('Population of City (10,000s)')
plt.ylabel('Profit ($10,000s)')
plt.title('Profit vs Population')
plt.grid(True)
plt.show()
```

```python
assert len(X1) == len(y1), "Mismatch between input and output sizes!"
m=len(y1)
print(f"Total valid training examples: {m}")
unique_X=sorted(set(X1))
unique_y=sorted(set(y1))

print(f"\nTotal unique values in X (Population): {len(unique_X)}")
print(unique_X)

print(f"\nTotal unique values in y (Profit): {len(unique_y)}")
print(unique_y)
```

```python

def compute_cost(theta0, theta1, X, y):
    m=len(y)
    error1=0#total error
    for i in range(m):
        y_hat=theta0 + theta1 * X[i]#getting prediction
        error=y_hat - y[i]
        error1 += error ** 2
    e=error1 / (2 * m)
    return e

def grad_descent(X, y, alpha, iter):
    theta0=0
    theta1=0
    m=len(y)
    t_cost=[]

    for it in range(iter):
        sum_errors_theta0=0
        sum_errors_theta1=0

        for i in range(m):
            h=theta0 + theta1 * X[i]
            error=h - y[i]
            sum_errors_theta0 += error
            sum_errors_theta1 += error * X[i]

        theta0=theta0 - (alpha / m) * sum_errors_theta0
        theta1=theta1 - (alpha / m) * sum_errors_theta1

        cost=compute_cost(theta0, theta1, X, y)
        t_cost.append(cost)

        #print(f"Iteration {it+1}: Cost={cost:.4f}, theta0={theta0:.4f}, theta1={theta1:.4f}")

    return theta0, theta1, t_cost
alpha=0.01
iter=1500

theta0,theta1,cost_history=grad_descent(X1, y1, alpha, iter)
print(f"Trained parameters: theta0 = {theta0:.4f}, theta1 = {theta1:.4f}")
plt.plot(X1, y1, 'rx', label='Training data')
plt.plot(X1, [theta0 + theta1 * xi for xi in X1], 'b-', label='Linear Regression')
plt.xlabel('Population of City (10,000s)')
plt.ylabel('Profit ($10,000s)')
plt.title('Linear Fit after Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
plt.plot(range(iter), cost_history, 'b-')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function Convergence')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

X=[]
y=[]

with open('ex1data2.txt', 'r') as file:
    for line in file:
        vals=line.strip().split(',')
        X.append([float(vals[0]), float(vals[1])])  # [Size, Bedrooms]
        y.append(float(vals[2]))                    # Price

m=len(y)
```

```python
assert len(X) == len(y), "Mismatch between input and output sizes!"
m=len(y)
print(f"Total valid training examples: {m}")
X_population_only=[row[1] for row in X]

unique_X=sorted(set(X_population_only))
unique_y=sorted(set(y))

print(f"\nTotal unique values in X (Population): {len(unique_X)}")
print(unique_X)

print(f"\nTotal unique values in y (Profit): {len(unique_y)}")
print(unique_y)
```

```python

def fea_normalize_np(X):
    mu=X.mean(axis=0)
    sigma=X.std(axis=0)
    X_norm=(X - mu) / sigma
    return X_norm, mu, sigma

def compute_cost_vectorized(X, y, theta):
    m=len(y)
    y_hat=X.dot(theta)#prediction
    error=y_hat - y
    return (1 / (2 * m)) * np.dot(error.T, error)

def grad_descent_vectorized(X,y,theta,alpha,iter):
    m=len(y)
    t_cost= []
    for _ in range(iter):
        y_hat=X.dot(theta)
        error=y_hat - y
        grad=(1 / m) * X.T.dot(error)
        theta=theta - alpha * grad
        t_cost.append(compute_cost_vectorized(X,y,theta))
    return theta, t_cost

X=np.array(X)
y=np.array(y)

X_norm,mu,sigma=fea_normalize_np(X)
X_prep=np.c_[np.ones(X_norm.shape[0]),X_norm] #adding bias col
lr=[0.01, 0.03, 0.1, 0.3, 1]
iter=50
theta_init=np.zeros(X_prep.shape[1])

plt.figure(figsize=(8, 5))
for alpha in lr:
    theta=theta_init.copy()
    _, costs= grad_descent_vectorized(X_prep,y,theta,alpha,iter)
    plt.plot(range(1, len(costs) + 1), costs,label=f'alpha = {alpha}')

plt.xlabel('Iterations')
plt.ylabel('Cost J(θ)')
plt.title('Cost vs. Iterations for Different Learning Rates')
plt.legend()
plt.grid(True)
plt.show()
#best is the one that gives fastest convergence and sharpest cost drop
#here 0.3 gave and became smooth under 10 iter
best_alpha= 0.3#based on plot
theta=theta_init.copy()
theta, _ =grad_descent_vectorized(X_prep, y, theta, best_alpha, 400)

x_pred=np.array([1650, 3])
x_pred_norm=(x_pred - mu) / sigma
x_pred_norm= np.insert(x_pred_norm, 0, 1)#adding bias
price= x_pred_norm.dot(theta)

print(f"\nPredicted price of a 1650 sq-ft, 3 br house (using gradient descent): ${price:.2f}")
```

```python
def normal_equation(X, y):
    #adding 1s of bias
    X_new=[]
    for i in range(len(X)):
        X_new.append([1]+X[i])


    #transpose
    #print(X_new[0])
    Xt=[]
    for i in range(len(X_new[0])):#3
        temp=[]
        for j in range(len(X_new)):
            temp.append(X_new[j][i])
        Xt.append(temp)

    #Xt * X
    XtX=[]
    for i in range(len(X_new[0])):  #i:row in Xt
     temp=[]
     for j in range(len(X_new[0])):  #j:column in X_new
        sum=0
        for k in range(len(X_new)):  #k:iterate over rows
            sum=sum+X_new[k][i] * X_new[k][j]
        temp.append(sum)
     XtX.append(temp)

    # Xt * y
    Xty=[]
    for i in range(len(X_new[0])):
        sum=0
        for j in range(len(X_new)):
            sum=sum+Xt[i][j]*y[j]
        Xty.append(sum)

    #Inverse of XtX
    thetas=list(np.linalg.inv(XtX).dot(Xty))
    return thetas

thetas=normal_equation(X, y)

price_ne = thetas[0] + thetas[1]*1650 + thetas[2]*3
print(f"Predicted price of a 1650 sq-ft, 3 br house (using normal equation): ${price_ne:.2f}")
```
